```python
import re
import pandas as pd

df = pd.read_csv("output_zh_optimized_dedup_part4.csv")


def extract_poj_or_original(text):
    match = re.search(r"\[poj\](.*?)\[/poj\]", str(text), re.IGNORECASE | re.DOTALL)
    if match:
        return match.group(1).strip()
    return text


# 新增一個欄位 chinese_text，根據規則處理
df["chinese_text"] = df["中文意譯"].apply(extract_poj_or_original)
df.to_csv("output_zh_optimized_dedup_part4.csv", index=False, encoding="utf-8-sig")
```
